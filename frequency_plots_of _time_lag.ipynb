```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame

# Compute the frequency of each unique combination of "Time Lag" and "Time lag sim"
frequency_data = data.groupby(['Time Lag', 'Time lag sim']).size().reset_index(name='Frequency')

# Plot the scatter plot with size based on frequency
sns.scatterplot(data=frequency_data, x="Time Lag", y="Time lag sim", size="Frequency", sizes=(20, 200))
plt.show()
```
