```python
data = data.drop(data.columns[0:11], axis=1)
```

```python
 from sklearn.linear_model import LinearRegression
data
```

```python
selected_columns = ['Interior Temperature','Time lag ', 'Unnamed: 21']
new_data = data[selected_columns].copy()
```

```python
new_data
```

```python
X = new_data.drop(['Unnamed: 21'], axis = 1)
Y = new_data['Unnamed: 21']
X.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
new_data[['Interior Temperature', 'Time lag ']]=scaler.fit_transform(new_data[['Interior Temperature', 'Time lag ']])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = reg.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
```
