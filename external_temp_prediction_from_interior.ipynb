```python
 from sklearn.linear_model import LinearRegression
data
```

```python
selected_columns = ['Inside Surface Temp','Time Lag', 'Corr T Ex']
new_data = data[selected_columns].copy()
```

```python
new_data
```

```python
X = new_data.drop(['Corr T Ex'], axis = 1)
Y = new_data['Corr T Ex']
X.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
new_data[['Inside Surface Temp','Time Lag']]=scaler.fit_transform(new_data[['Inside Surface Temp','Time Lag']])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred = reg.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
```
